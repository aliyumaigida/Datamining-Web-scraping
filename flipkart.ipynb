```python
product_image = []
product_name = []
Original_price = []
Percentage = []
Discount_price = []
Color = []
RatingReview = []
Promo = []

for content in prod_contents:
    # if isinstance(content, str):
    #     content = BeautifulSoup(content, "html.parser")  # Ensure parsing
    
    for product in content:
        # if isinstance(product, str):
        #     continue  # Skip strings, only process BeautifulSoup objects

         # Scrape the product name
        image = product.findAll("div", {"class": "_4WELSP WH5SS-"})
        if len(image) == 0:
            product_image.append(np.nan)
        else:
               product_image.append(image[0].findAll("img",{"loading":"eager"})[0]['src'])
 

        # Scrape the product name
        name = product.findAll("a", {"class": "wjcEIp"})
        if len(name) == 0:
            product_name.append(np.nan)
        else:
            product_name.append(name[0].text.strip())

        # Scrape the product original_price
        price= product.findAll("div", {"class": "yRaY8j"})
        if len(price) == 0:
            Original_price.append(np.nan)
        else:
            Original_price.append(price[0].text.strip())

               # Scrape the product percentage
        Per = product.findAll("div", {"class": "UkUFwK"})
        if len(Per) == 0:
            Percentage.append(np.nan)
        else:
            Percentage.append(Per[0].text.strip())

               # Scrape the product discount-price
        Discount = product.findAll("div", {"class": "Nx9bqj"})
        if len(Discount) == 0:
            Discount_price.append(np.nan)
        else:
            Discount_price.append(Discount[0].text.strip())

              # Scrape the product color
        Col= product.findAll("div", {"class": "NqpwHC"})
        if len(Col) == 0:
           Color.append(np.nan)
        else:
            Color.append(Col[0].text.strip())

               # Scrape the product rating/review
        Rating = product.findAll("div", {"class": "_5OesEi afFzxY"})
        if len(Rating) == 0:
           RatingReview.append(np.nan)
        else:
            RatingReview.append(Rating[0].text.strip())

              # Scrape the product promo
        Pro = product.findAll("div", {"class": "n5vj9c"})
        if len(Pro) == 0:
           Promo.append(np.nan)
        else:
            Promo.append(Pro[0].text.strip())

# Debugging - Ensure lists have equal lengths
print(f"product_image: {len(product_image)}")
print(f"Products Name: {len(product_name)}")
print(f"Original_price: {len(Original_price)}")
print(f"Percentage: {len(Percentage)}")
print(f"Discount_price: {len(Discount_price)}")
print(f"Color: {len(Color)}")
print(f"RatingReview: {len(RatingReview)}")
print(f"Promo: {len(Promo)}")
```
